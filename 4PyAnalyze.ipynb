Find all instances of bee names ending in 'en' 'ed' or with a '-' using perl regular expression. A RegEx, or Regular Expression, is a sequence of characters that forms a search pattern.

RegEx can be used to check if a string contains the specified search pattern.

```python
import pandas as pd
import re

# Read the scientific and common name lookup csv file into a DataFrame
df3=pd.read_csv('/workspaces/myfolder/SASPythonDataScientists/Bumblebee_Others_Scientific_Common_Names.csv' , encoding='latin-1')
```

```python
df3.info()
```

DataFrames can be filtered in multiple ways; the most intuitive of which boolean indexing creating a series of True/False values.

```python
# Regular expression pattern
pattern = r'((ed|ern)\W)|-'

# Subset the DataFrame based on the regex pattern
df_regex = df3[df3['CommonName'].str.contains(pattern, regex=True)]

# Display the filtered DataFrame
print(df_regex)
```

```python
df_regex.shape
```

Detailed Explanation
df3['CommonName'].str.contains(pattern, regex=True): This function is used to generate a Boolean mask based on whether each entry in the commonname column matches the pattern.

df3['CommonName']: This selects the commonname column from the DataFrame df3.
.str.contains(pattern, regex=True): This method checks each entry in the commonname column to see if it contains a substring that matches the regular expression pattern(defined earlier).
regex=True: This argument ensures that the pattern is treated as a regular expression. It‚Äôs the default value, so it‚Äôs optional but included here for clarity.
df3[...]:
The Boolean Series returned by str.contains() (which is True or False for each row) is used to filter df3.
df3[df3['commonname'].str.contains(pattern, regex=True)]: filters the DataFrame df3 to include only the rows where the commonname matches the regex pattern.
