```python
def dollar_to_num(value):
    return float(value.replace('$', '').replace(',', '').replace(' ',''))
```

```python
df['actual_num'] = df['ACTUAL'].apply(dollar_to_num)
df['predict_num'] = df['PREDICT'].apply(dollar_to_num)
 
```
